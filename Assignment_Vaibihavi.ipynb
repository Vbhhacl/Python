1. Move All Zeros to End Problem: Given an integer array, move all 0s
to the end while maintaining the relative order of non-zero
elements. Example: Input: [0,1,0,3,12] Output: [1,3,12,0,0]
Constraints: 1 ≤ n ≤ 10^5

```python
num = [int(x) for x in input("Enter elements: ").split()]
pos = 0  
for i in range(len(num)):
    if num[i] != 0:
        num[pos], num[i] = num[i], num[pos]
        pos += 1
print("Result:", num)
```

2. Longest Consecutive Ones Problem: Given a binary array, return the
maximum number of consecutive 1s. Example: Input: [1,1,0,1,1,1]
Output: 3 Constraints: 1 ≤ n ≤ 10^5

```python
num = [int(x) for x in input("Enter binary elements: ").split()]
count = 0
i = 0
for x in num:
    if x == 1:
        i += 1
        if i > count: 
            count = i
    else:
        i = 0
print("Max consecutive 1s:", count)
```

3. Majority Element Problem: Find the element that appears more than
⌊n/2⌋ times. Example: Input: [2,2,1,1,1,2,2] Output: 2 Constraints:
n ≥ 1

```python
num = [int(x) for x in input("Enter elements: ").split()]
n = len(num)
ans = 0
for x in set(num):
    if num.count(x) > n // 2:
        ans = x
        break
print("Majority Element:", ans)
```

4. Second Largest Element Problem: Find the second largest element in
array. Example: Input: [10,20,4,45,99] Output: 45 Constraints: n ≥ 2

```python
num = [int(x) for x in input("Enter elements: ").split()]

num.sort()
print("Second Largest Element:", num[-2])
```

5. Fibonacci Number Problem: Return nth Fibonacci number (0-indexed).
Example: Input: 6 Output: 8 Constraints: 0 ≤ n ≤ 50

```python
n = int(input("Enter n: "))
a = 0
b = 1
for i in range(n):
    temp = a + b
    a = b
    b = temp
print("Result:", a)
```

6. Frequency of Characters Problem: Count frequency of each character
in string. Example: Input: “aabbc” Output: {‘a’:2,‘b’:2,‘c’:1}
Constraints: 1 ≤ len(s) ≤ 10^5

```python
s = input("Enter string: ")
freq = {}
for char in s:
    if char in freq:
        freq[char] = freq[char] + 1
    else:
        freq[char] = 1
print("Result:", freq)
```

7. Rotate Array by K Problem: Rotate array to right by k steps.
Example: Input: [1,2,3,4,5,6,7], k=3 Output: [5,6,7,1,2,3,4]
Constraints: 1 ≤ n ≤ 10^5

```python
num = [int(x) for x in input("Enter elements: ").split()]
k = int(input("Enter k: "))
n = len(num)
k = k % n  
result = num[n-k:] + num[:n-k]
print("Result:", result)
```

8. Find Missing Number Problem: Given array of size n containing
numbers from 0 to n, find missing number. Example: Input: [3,0,1]
Output: 2 Constraints: Exactly one number missing

```python
nums = [int(x) for x in input("Enter elements: ").split()]
n = len(nums)
total_sum = n * (n + 1) // 2
actual_sum = sum(nums)
print("Result:", total_sum - actual_sum)
```

9. Balanced Parentheses Problem: Check if parentheses string is valid.
Example: Input: “()” Output: True Constraints: 1 ≤ len(s) ≤ 10^5

```python
s = input("Enter parentheses: ")
while "()" in s:
    s = s.replace("()", "")
if s == "":
    print("True")
else:
    print("False")
```

10. Reverse Words in a String Problem: Reverse the order of words in a
string. Example: Input: “the sky is blue” Output: “blue is sky the”
Constraints: 1 ≤ len(s) ≤ 10^4

```python
words = input("Enter string: ").split()
words.reverse()
result = " ".join(words)
print("Result:", result)
```

11. Find All Duplicates Problem: Return all elements that appear twice
in array. Example: Input: [4,3,2,7,8,2,3,1] Output: [2,3]
Constraints: 1 ≤ n ≤ 10^5

```python
n = [int(x) for x in input("Enter numbers: ").split()]
for x in set(n):
    if n.count(x) > 1:
        print("Duplicate found:", x)
```

12. Subarray with Given Sum Problem: Find if there exists a subarray
with sum equal to target (positive numbers only). Example: Input:
[1,4,20,3,10,5], target=33 Output: True Constraints: 1 ≤ n ≤ 10^5

```python
n = [int(x) for x in input("Enter numbers: ").split()]
t = int(input("Enter target: "))
ans = False
for i in range(len(n)):
    s = 0 
    for j in range(i, len(n)):
        s = s + n[j]
        if s == t:
            ans = True
print("Result:", ans)
```
